```python
import csv
dept_salary = {} 
with open("data.csv","r") as file:
    reader = csv.reader(file)
    for name, dept, sal in reader: 
        sal = float(sal)
        dept_salary.setdefault(dept, []).append(sal)

print("average salary by Department:")
for dept, sals in dept_salary.items ():
    print(f"\t{dept}: {sum(sals)/len(sals):.2f}")
```
