# Index and prints the element at that index from a predefined list.

```python
# Predefined list
numbers = [10, 20, 30, 40, 50]

try:
    # Ask user for index
    index = int(input("Enter an index: "))
    
    # Print element at given index
    print("Element at index", index, "is:", numbers[index])

except Exception as e:
    # Print specific exception message
    print("Error:", e)
```
