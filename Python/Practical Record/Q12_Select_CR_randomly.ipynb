# Select CR randomly & shuffle student list

```python
import random

students = ["Aman", "Lokesh", "Leeladhar", "Kunal", "Mukesh", "Geeradhar"]

print("students list:", students)
```

```python
# Select CR randomly
cr = random.choice(students)
print("Class Representative (CR):", cr)

# Shuffle student list
random.shuffle(students)
print("Shuffled Student List:", students)
```
