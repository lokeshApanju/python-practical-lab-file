# Read a sentence and display how many times each word appears

```python
import string

text = input("Enter text: ").lower()

print("Text is : ",text)
```

```python
for ch in string.punctuation:
    text = text.replace(ch, "")

words = text.split()
freq = {}

for word in words:
    freq[word] = freq.get(word, 0) + 1

print(freq)
```
