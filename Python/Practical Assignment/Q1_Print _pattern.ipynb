```python
print("(a)")
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```

```python
print("(b)")
n = 4
for i in range(1, n + 1):
    print("  " * (n - i), end="")
    print("* " * (2 * i - 1))
```

```python
print("(c)")
n = 3
# Upper half
for i in range(1, n + 1):
    print("  " * (n - i), end="")
    print("* " * (2 * i - 1))
# Lower half
for i in range(n - 1, 0, -1):
    print("  " * (n - i), end="")
    print("* " * (2 * i - 1))
```

```python
print("(d)")
rows = 4
cols = 3
for i in range(rows):
    for j in range(cols):
        if i == 0 or i == rows - 1 or j == 0 or j == cols - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
