# list of employee records as dictionaries, sorting  by salary and display the sorted list.

```python
employees = [
    {"name": 'Ram', "salary": 45000},
    {"name": 'Shyam', "salary": 52000},
    {"name": 'Shiv', "salary": 60000},
    {"name": 'Mohan', "salary": 48000},
    {"name": 'Sohan', "salary": 75000},
    {"name": 'Ramesh', "salary": 30000},
]

sorted_employees = sorted(employees, key=lambda x: x['salary'])
for emp in sorted_employees:
    print(f"Name: {emp['name']}, Salary: {emp['salary']}")
```
