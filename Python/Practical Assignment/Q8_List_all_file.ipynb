# All files in a directory and categorizes them by file extension.

```python
files = ['report.pdf', 'data.csv', 'photo.jpg', 'summary.pdf', 'script.py']
file_dict = {}
print("Files is : ",files)
```

```python
for file in files:
    ext = file.split('.')[-1]
    file_dict.setdefault(ext, []).append(file)

print("categorizes by file extension.",file_dict)
```
