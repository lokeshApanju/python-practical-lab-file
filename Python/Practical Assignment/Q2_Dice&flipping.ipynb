```python
import random

print("1. Dice  2. Coin")
choice = input("Choose: ")
print(f"You choose : {choice}")
```

```python
if choice == "1":
    num = int(input("Enter number (1-6): "))
    count = sum(random.randint(1,6) == num for _ in range(10000))
    print("Enter number (1-6):", num)
    print(f"Probability: {count/10000:.3f}")
else:
    side = input("Heads or Tails? ").lower()
    count = sum(random.choice(['H','T']) == 'H' for _ in range(10000))
    if side == "tails":
        count = 10000 - count
    print("Enter Heads or Tails?:", side)
    print(f"Probability: {count/10000:.3f}")
```
